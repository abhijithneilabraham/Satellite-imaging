```

import os
import folium
```

```
import numpy as np

data = (
    np.random.normal(size=(100, 3)) *
    np.array([[1, 1, 1]]) +
    np.array([[48, 5, 1]])
).tolist()
```

```
from folium.plugins import HeatMap

m = folium.Map(location=[8.5459, 76.9063], tiles='stamentoner', zoom_start=12)

HeatMap(data).add_to(m)



m
```
